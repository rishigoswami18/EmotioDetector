```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/train.txt',sep=';',header=None,names=['text','emotions'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
unique_emotions = df['emotions'].unique()
emotion_numbers ={}
i = 0
for emotion in unique_emotions:
    emotion_numbers[emotion] = i
    i+=1
df['emotions'] = df['emotions'].map(emotion_numbers)    
```

```python
df
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
import string 
def remove_punctuation(text):
    return text.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punctuation)
```

```python
def remove_numbers(txt):
    new = ""
    for char in txt:
        if not char.isdigit():
            new += char
    return new
df['text'] = df['text'].apply(remove_numbers)    
```

```python
def remove_emojis(txt):
    new = ""
    for char in txt:
        if char.isascii():
            new += char
    return new
df['text'] = df['text'].apply(remove_emojis)    
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stops = set(stopwords.words('english'))
stops
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
    new = []
    for word in txt.split():
        if word not in stops:
           new.append(word)
    return ' '.join(new)
```

```python
df['text'] = df['text'].apply(remove)
```

```python
df.loc[1]['text']
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df['text'],df['emotions'],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
bow_vectorizer = CountVectorizer()
```

```python
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_train_bow
```

```python
X_test_bow = bow_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score
```

```python
nb_model = MultinomialNB()
```

```python
nb_model.fit(X_train_bow,y_train)
```

```python
pred_bow = nb_model.predict(X_test_bow)
print(accuracy_score(y_test,pred_bow))
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf,y_train)
```

```python
y_pred = nb2_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
logistic_model = LogisticRegression(class_weight="balanced",max_iter=1000)
```

```python
logistic_model.fit(X_train_tfidf,y_train)
```

```python
log_pred = logistic_model.predict(X_test_tfidf)
```

```python
accuracy_score(y_test,log_pred)
```

```python
import pandas as pd
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Load dataset
df = pd.read_csv("data/train.txt", sep=";", header=None, names=["text", "emotion"])

# Features & labels
X = df["text"]
y = df["emotion"]

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)
X_vec = vectorizer.fit_transform(X)

# Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_vec, y)

# Save model and vectorizer
joblib.dump(model, "model.pkl")
joblib.dump(vectorizer, "vectorizer.pkl")

print("✅ Model and vectorizer saved successfully")

```
